## 여러 엑셀 파일

```python
NC시즌 = []

for 시즌 in range(2013, 2016):
    경로 = 'data/NC Dinos {}.xlsx'.format(시즌)
    프레임 = office.Excel(경로)
    프레임['시즌'] = 시즌
    프레임.name = 'NC {}'.format(시즌)
    NC시즌.append(프레임)
```

```python
NC시즌[0][:3]
```

```python
NC시즌[0].name
```

```python
office.to_excel(NC시즌, 'NC시즌.xlsx')
```

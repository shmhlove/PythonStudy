#### 자동 구글 검색

검색을 수행해 상위 5개 결과의 링크를 웹브라우저에서 열기

```python
def 구글검색(키워드, 탭=5):
    url = 'http://google.com/search?q={0}'.format(키워드)
    result_web_page = web.parse_html(url)
    
    결과링크 = result_web_page.select('.r a')
    for 링크 in 결과링크[:탭]:
        webbrowser.open('http://google.com' + 링크.get('href'))
```

```python
구글검색('칼퇴근 맛집')
```
